# Sales Forecasting with Multiple Business Datasets

This notebook uses the available business datasets to build a sales forecast model and validate predictions.

```
import pandas as pdimport numpy as npimport matplotlib.pyplot as pltfrom sklearn.ensemble import RandomForestRegressorfrom sklearn.metrics import mean_squared_error, mean_absolute_percentage_errorimport os%matplotlib inline
```

```
repo_root = os.path.abspath(os.path.join('..'))data_root = os.path.abspath(os.path.join(repo_root, '..'))train = pd.read_csv(os.path.join(data_root, 'train.csv'), parse_dates=['date'])stores = pd.read_csv(os.path.join(data_root, 'stores.csv'))transactions = pd.read_csv(os.path.join(data_root, 'transactions.csv'), parse_dates=['date'])holidays = pd.read_csv(os.path.join(data_root, 'holidays_events.csv'), parse_dates=['date'])oil = pd.read_csv(os.path.join(data_root, 'oil.csv'), parse_dates=['date'])train = train.groupby(['date', 'store_nbr'], as_index=False)['sales'].sum()transactions = transactions.groupby(['date', 'store_nbr'], as_index=False)['transactions'].sum()data = train.merge(transactions, on=['date', 'store_nbr'], how='left')data = data.merge(stores[['store_nbr', 'city', 'state', 'type', 'cluster']], on='store_nbr', how='left')oil = oil.fillna(method='ffill').fillna(method='bfill')data = data.merge(oil, on='date', how='left')national = holidays[holidays.locale == 'National'][['date']].assign(national_holiday=1)regional = holidays[holidays.locale == 'Regional'][['date', 'locale_name']].rename(columns={'locale_name': 'state'}).assign(regional_holiday=1)local = holidays[holidays.locale == 'Local'][['date', 'locale_name']].rename(columns={'locale_name': 'city'}).assign(local_holiday=1)data = data.merge(national, on='date', how='left')data = data.merge(regional, on=['date', 'state'], how='left')data = data.merge(local, on=['date', 'city'], how='left')for col in ['national_holiday', 'regional_holiday', 'local_holiday']:    data[col] = data[col].fillna(0)data['dayofweek'] = data.date.dt.dayofweekdata['month'] = data.date.dt.monthdata['weekofyear'] = data.date.dt.isocalendar().week.astype(int)data['day'] = data.date.dt.daydata['year'] = data.date.dt.yeardata['is_weekend'] = (data.dayofweek >= 5).astype(int)data = data.sort_values(['store_nbr', 'date'])for lag in [7, 14, 28]:    data[f'lag_{lag}'] = data.groupby('store_nbr')['sales'].shift(lag)for window in [7, 14]:    data[f'rolling_{window}'] = data.groupby('store_nbr')['sales'].shift(1).rolling(window).mean().reset_index(level=0, drop=True)data = pd.get_dummies(data, columns=['type'], prefix='type')data = data.fillna(0)data.head()
```

```
cutoff = data.date.max() - pd.Timedelta(days=60)train_df = data[data.date <= cutoff]val_df = data[data.date > cutoff]feature_cols = ['transactions', 'dcoilwtico', 'national_holiday', 'regional_holiday', 'local_holiday', 'dayofweek', 'month', 'weekofyear', 'day', 'year', 'is_weekend', 'cluster', 'lag_7', 'lag_14', 'lag_28', 'rolling_7', 'rolling_14'] + [c for c in data.columns if c.startswith('type_')]train_df = train_df.dropna(subset=feature_cols + ['sales'])val_df = val_df.dropna(subset=feature_cols + ['sales'])X_train = train_df[feature_cols]y_train = train_df['sales']X_val = val_df[feature_cols]y_val = val_df['sales']model = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=42)model.fit(X_train, y_train)pred = model.predict(X_val)rmse = np.sqrt(mean_squared_error(y_val, pred))mape = mean_absolute_percentage_error(y_val, pred)rmse, mape
```
